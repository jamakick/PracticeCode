Import pandas and read in the csv file with the index_col being 0 so our id acts as the index

```python
import pandas as pd

birth = pd.read_csv("birthweight.csv", index_col=0)
print(birth)
```

Use birth.describe() to get the numeric information for every column. We then print out the min, max, and range of the mother ages and father ages.

```python
desc = birth.describe()

print("Minimum age of mothers is", desc.motherage['min'])
print("Maximum age of mothers is", desc.motherage['max'])
print("Range of ages for mothers is ", int(desc.motherage['max']) - int(desc.motherage['min']))
print()
print("Minimum age of fathers is", desc.fage['min'])
print("Maximum age of fathers is", desc.fage['max'])
print("Range of ages for fathers is ", int(desc.fage['max']) - int(desc.fage['min']))
print()
print(desc)
```

Find the counts of each value in the smoker series from our dataframe. There are more smokers than non smokers, 22 to 18

```python
smoke = birth['smoker'].value_counts()
print(smoke)
```

Since our smoker values should only be 1.0 or 0.0, we can check the values to see if they are inside of that list. If they aren't then the data is removed

```python
birth = birth[birth.smoker.isin([0.0,1.0])]

print(birth)
```

Sorted the column names alphabetically with sort index at axis 1

```python
birth = birth.sort_index(axis=1)
print(birth)
```

Use sort_values to show the birthweights in descending order, and use iloc to display the heaviest baby

```python
weight = birth[["Birthweight"]].sort_values(by="Birthweight", ascending = False)

print(weight.iloc[0])
```

Gestation and Length have the highest positive correlation with each other, of .626. This is found from using the rank() function on a dataframe with all 3 columns and then using the .corr function with the spearman method

```python
correlateBirth = birth[["Gestation", "headcirumference", "length"]].rank().corr(method="spearman")

print(correlateBirth)
```
