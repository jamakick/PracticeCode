Import pandas and use the read_csv function to read our csv file into a pandas Dataframe.

```python
import pandas as pd

tobacco = pd.read_csv("tobacco.csv")
```

Use the dtypes attribute to print the type of all the columns, and use the shape attribute to show the shape of the Dataframe.

```python
print(tobacco.dtypes)
print(tobacco.shape)
```

Use boolean indexing on our tobacco Dataframe to find any location abbreviations that match IN, meaning Indiana. I then save this new dataframe to a variable tobacco_in for future use.

```python
tobacco_in = tobacco[tobacco['LocationAbbr'] == "IN"]

print(tobacco_in)
```

Use fancy indexing on our tobacco_in dataframe to show only the columns included in our given list.

```python
print(tobacco_in[['YEAR', "Gender", "Age", "Data_Value", "Sample_Size"]])
```

Here we stack our fancy indexing and boolean indexing together in one line.

```python
indexedTobacco = tobacco[tobacco["LocationAbbr"] == "IN"][['YEAR', "Gender", "Age", "Data_Value", "Sample_Size"]]
print(indexedTobacco)
```

Here I save all the columns from our dataframe into a variable. I then loop through the indexes 1 to 6 and delete the column from columns associated with that negative index from the Dataframe.

```python
columns = tobacco_in.columns

for i in range(1,7):
    del tobacco_in[columns[-i]]

print(tobacco_in)
```

I create a new Dataframe called tobaccoFiltered, and use my original tobacco Dataframe to index. I stack 3 boolean indexes onto it, Age being the string 20 and older, the locationabbr being IN, and the Year being 2016. I then use my fancy indexing to get only the columns I want in the next line. There was one two high school graudates, one being in college, and one non graduate, So high school graduates had more smokers in this case.

```python
tobaccoFiltered = tobacco[tobacco["Age"] == "Age 20 and Older"][tobacco['LocationAbbr'] == "IN"][tobacco["YEAR"] == "2016"]

tobaccoFiltered = tobaccoFiltered[["YEAR", "LocationDesc", "TopicDesc", "MeasureDesc", "Data_Value", "Education"]]

print(tobaccoFiltered)
```
