Import pandas and matplotlib
Read csv in and create a dataframe of the sorted median incomes, slicing off the top 10
give this dataframe to the plot command to make the shown graph

```python
import pandas as pd, matplotlib.pyplot as plt

contents = pd.read_csv("recent-grads.csv")

medianContents = contents[["Major", "Median"]].sort_values(by="Median", ascending = False)[:10]


medianContents.plot(x = "Major", kind = "barh", y = "Median", grid = True, title = "Median income by Major", legend = True)

plt.show()

```

Read in our weather csv file
to find the threshold of 80% missing values, we take the len of the dataframe, divide it by 10 and then multiply by 8, rounding it up to get the threshold where we cut off a column

We then look at every column to see if it has an M in it's value counts since that means missing according to the weather website
We then check if the count of M's is higher or equal to the threshold, and if it is we drop that column from our dataframe

```python
weather = pd.read_csv("weather.csv")

weathercol = weather.columns
weatherlen = len(weather)
missingThreshold  = round((weatherlen / 10) * 8)

for col in weathercol:
    if "M" in weather[col].value_counts():
        if weather[col].value_counts()["M"] >= missingThreshold:
            weather = weather.drop(columns=col)
            
            
print(weather)
```

Here we check to see if any rows in the WTR column have a T, meaning a trace amount of water. We then change it to 0.001 per instructions using the .at attribute

```python
for x in range(len(weather["WTR"])):
    if weather["WTR"][x] == "T":
        weather.at[x, "WTR"] = 0.001
        
print(weather)
```
